### Exercise 1. 

Read Total Profit for all months and show it using a line plot with Month Number as your baseline

Total profit data provided for each month. Generated line plot must include the following properties:

1. X label name = Month Number
2. Y label name = Total profit

````python
# Exercise 1
````

````python
#matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd
````

````python
company_sales = pd.read_csv("data/company_sales_data.csv")
fig = plt.figure()
ax = plt.axes()
plt.plot(company_sales.month_number, company_sales.total_profit)
ax.set(xlim=(1, 12), ylim=(150000, 450000),
       xlabel='Month Number', ylabel='Total profit',
       title='Total Profit of All months');
````

### Exercise 3.

Read all product sales data and show it using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).

````python
# facecream 	facewash 	toothpaste 	bathingsoap 	shampoo 	moisturizer 	
fig = plt.figure()
ax = plt.axes()
facecream, = plt.plot(company_sales.month_number, company_sales.facecream, ':k')

facewash, = plt.plot(company_sales.month_number, company_sales.facewash, '--b')

toothpaste, = plt.plot(company_sales.month_number, company_sales.toothpaste, '-r')

bathingsoap, = plt.plot(company_sales.month_number, company_sales.bathingsoap, '-.g')

shampoo, = plt.plot(company_sales.month_number, company_sales.shampoo, '-om')

moist, = plt.plot(company_sales.month_number, company_sales.moisturizer, ':vy')

plt.legend([facecream, facewash, toothpaste, bathingsoap, shampoo, moist], 
               ["Facecream", "Facewash", "Toothpaste", "Bathing Soap", "Shampoo", "Moisturizer"], loc="upper left")

ax.set(xlim=(0, 12), ylim=(0, 15000),
       xlabel='Month Number', ylabel='Solid units number',
       title='All Product Sales')
````

### Exercise 5.

Read face cream and face wash product sales data and show it using the bar chart

The bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

````python

monthList  = company_sales ['month_number'].tolist()
faceCremSalesData   = company_sales ['facecream'].tolist()
faceWashSalesData   = company_sales ['facewash'].tolist()

plt.bar([a-0.25 for a in monthList], faceWashSalesData, width= 0.25, color = 'blue', label = 'Facewash', align='edge')
plt.bar([a+0.25 for a in monthList], faceCremSalesData, width= -0.25, color = 'yellow', label = 'Facecream', align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(monthList)
plt.grid(True, linewidth= 1)
plt.title('Facewash and facecream sales data')
plt.show()
````

### Exercise 6.
Write a script that creates your favorite visual type you have learned about, and display the visual once from within the notebook:

````python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

# Data for three-dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
````
